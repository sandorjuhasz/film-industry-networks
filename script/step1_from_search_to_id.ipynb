```python
url = 'http://mandarchiv.hu/tart/jatekfilm?action=search&title=&stab=&szereplo=&gyarto=&szinesseg=&gyartas_tol=2009&gyartas_ig=2010&tartalom=&dij='
response = requests.get(url, headers=get_header(USER_AGENTS))
soup = BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
link = []
```

```python
input_link = soup.findAll('a', {'class': 'title'})
for l in input_link:
    l_temp = l.get("href")
    l_temp = l_temp[-9:]
    link.append(l_temp)
```

```python
link = {'link' : pd.Series(link)}
link_2009_2010 = pd.DataFrame(link)
```
